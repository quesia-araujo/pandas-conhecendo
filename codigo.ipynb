## Qual o valor médio de aluguel por tipo de imóvel?


```python
dados['Valor'].mean() # Média dos valores da coluna Valor
```

```python
dados.groupby('Tipo').mean(numeric_only=True) # Média dos valores agrupados por Tipo
```

```python
dados.groupby('Tipo')['Valor'].mean() # Média dos valores agrupados por tipo e retornando apenas a coluna Valor
```

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values(by='Valor', ascending=False) # Média dos valores agrupados por tipo, retornando apenas a coluna Valor e ordenando em ordem decrescente
```

```python
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values(by='Valor', ascending=False)
df_preco_tipo.plot(kind='barh', figsize=(10,5), color='purple') # Gráfico de barras horizontais
```
